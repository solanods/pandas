```python
def contarFaltas(dataframe):
  # Obter lista com nome de todos operadores
  operadores = []
  for linha in dataframe.index:
    if search ('[A-Z]+', str(dataframe['Unnamed: 3'][linha])):
      operadores.append(dataframe['Unnamed: 3'][linha])

  operadores = [operador for operador in operadores if operador.isupper()]

  # Criar dataframe com as colunas de dados alvo
  nome_missao = pd.concat([dataframe['Unnamed: 3'], dataframe['Unnamed: 8']], axis=1, ignore_index=False)

  # Obter as listas de missoes varrendo a coluna com um range
  # O valor no range deve ser igual ao número de operadores
  i = 8
  j =39
  missoes = []
  for x in range(len(operadores)):
    missoes.append(nome_missao['Unnamed: 8'].iloc[i:j].to_list())
    i += 66
    j += 66
  
  # Contador de faltas
  FAM = []
  FAT = []
  FAN = []
  for missao in missoes:
    FAM.append(missao.count('FAM'))
    FAT.append(missao.count('FAT'))
    FAN.append(missao.count('FAN'))

  # Zip the columns
  resultado = pd.DataFrame(list(zip(operadores, FAM, FAT, FAN)), columns = ['Operador', 'FAM', 'FAT', 'FAN'])
  resultado['Ausências'] = resultado['FAM'] + resultado['FAT'] + resultado['FAN']
  resultado['Mês'] = dataframe['Unnamed: 12'][3]

  return resultado.sort_values(by=['Ausências'], ascending=False)
```

```python
for mes in COD_PRMS:
  print(contarFaltas(PRMS[mes]))
```
