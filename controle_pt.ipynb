```python
# Preenche nan
df1 = df1.fillna('-')
```

```python
# Converte os valores de int para string
data_string = []
for i in df1['Unnamed: 2']:
  data_string.append(str(i))
```

```python
# Adiciona 0 ao início de cada string de 7 caracteres
nova_data = []
for item in data_string:
  if len(item) == 7:
    nova_data.append("0" + item)
  else:
    nova_data.append(item) 
```

```python
df1['DATA'] = nova_data
```

TRATAMENTO DA COLUNA DATA

```python
# Verificar o padrão data. Manter somente entradas de 8 dígitos
controlept['DATA'].value_counts()
```

```python
# Nesse caso usaremos uma expressão regular
padrao_data = r'\d{8}'
mascara_data = controlept['DATA'].str.contains(padrao_data, case=False, na=False)
controlept = controlept[mascara_data]
```

```python
controlept['DATA'].value_counts()
```

```python
# Indices quadruplicados
controlept['DATA'][5]
```

```python
# Normaliza os indices
controlept = controlept.reset_index(drop=True)
```

```python
controlept['DATA'][5]
```

```python
# Ao normalizar as datas corrigir as exceções
from datetime import datetime
parser_data = []
for data in controlept['DATA']:
  try:
    novadata = datetime.strptime(data, '%d%m%Y').strftime('%d/%m/%Y')
    parser_data.append(novadata)
  
  # Criada data ficticia para posterior exclusão
  except:
    parser_data.append('20/09/2027')
```

```python
controlept['DATE'] = parser_data
```

```python
controlept
```

```python
parser_turno = []
for t in controlept['TURNO']:
  t = str(t).upper()
  parser_turno.append(t)
```

```python
controlept['P_TURNO'] = parser_turno
```

```python
controlept['P_TURNO'].value_counts()
```
